# Sample ADB Data Pipeline Notebook

This notebook demonstrates a simple Azure Databricks pipeline workflow using a sample dataset.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("sample_datapipeline").getOrCreate()

data = [
    (1, 10.0),
    (2, 20.5),
    (3, 15.75),
]
columns = ["id", "value"]
df = spark.createDataFrame(data, columns)

result = df.withColumn("scaled_value", col("value") * 1.1)
result.show()

output_path = "/tmp/adb_pipeline_output"
result.write.mode("overwrite").parquet(output_path)

print("Pipeline completed. Output written to:", output_path)
```
